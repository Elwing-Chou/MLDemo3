```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris["data"], columns=iris['feature_names'])
df["target"] = iris["target"]
df.to_csv("iris.csv", encoding="utf-8", index=False)
df
```

```python
from sklearn.model_selection import train_test_split
# 90%題目, 10%題目, 90%答案, 10%答案
x_train, x_test, y_train, y_test = train_test_split(iris["data"], iris["target"], test_size=0.1)
```

```python
from sklearn.cluster import KMeans
clu = KMeans(n_clusters=3)
clu.fit(x_train)
```

```python
clu.labels_
```

```python
pre = clu.predict(x_test)
from sklearn.metrics import accuracy_score
print("預測:", list(pre))
print("正確:", list(y_test))
print("正確率:", accuracy_score(pre, y_test))
```

```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
# 如果我不知道, 我就會選K=2
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
scores = []
xs = []
# range(2, 6) -> [2, 3, 4, 5]
for n in range(2, 6):
    clu = KMeans(n_clusters=n)
    clu.fit(iris["data"])
    s = silhouette_score(iris["data"], clu.labels_)
    scores.append(s)
    xs.append(n)
plt.title("Sihouette Score")
plt.plot(xs, scores, "mo--")
plt.show()
```
