```python
# http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz
import glob
import pandas as pd
traindf = pd.DataFrame(columns=["content", "sentiment"])
for fn in glob.glob("aclImdb/train/neg/*.txt"):
    # f = open(fn, "r", encoding="utf-8")
    with open(fn, "r", encoding="utf-8") as f:
        article = f.read()
    s = pd.Series([article, 0], index=["content", "sentiment"])
    traindf = traindf.append(s, ignore_index=True)
for fn in glob.glob("aclImdb/train/pos/*.txt"):
    # f = open(fn, "r", encoding="utf-8")
    with open(fn, "r", encoding="utf-8") as f:
        article = f.read()
    s = pd.Series([article, 1], index=["content", "sentiment"])
    traindf = traindf.append(s, ignore_index=True)
traindf
```

```python
testdf = pd.DataFrame(columns=["content", "sentiment"])
for fn in glob.glob("aclImdb/test/neg/*.txt"):
    # f = open(fn, "r", encoding="utf-8")
    with open(fn, "r", encoding="utf-8") as f:
        article = f.read()
    s = pd.Series([article, 0], index=["content", "sentiment"])
    testdf = testdf.append(s, ignore_index=True)
for fn in glob.glob("aclImdb/test/pos/*.txt"):
    # f = open(fn, "r", encoding="utf-8")
    with open(fn, "r", encoding="utf-8") as f:
        article = f.read()
    s = pd.Series([article, 1], index=["content", "sentiment"])
    testdf = testdf.append(s, ignore_index=True)
testdf
```

```python
from keras.preprocessing.text import Tokenizer
tok = Tokenizer(num_words=2000)
tok.fit_on_texts(traindf["content"])
tok.word_index
```

```python
traindf_seq = tok.texts_to_sequences(traindf["content"])
testdf_seq = tok.texts_to_sequences(testdf["content"])
pd.DataFrame(traindf_seq)
```

```python
from keras.preprocessing import sequence
traindf_pad = sequence.pad_sequences(traindf_seq, maxlen=100)
testdf_pad = sequence.pad_sequences(testdf_seq, maxlen=100)
pd.DataFrame(testdf_pad)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout, Embedding
model = Sequential()
model.add(Embedding(input_dim=2000, output_dim=32, input_length=100))
model.add(Flatten())
model.add(Dense(units=256, activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(units=1, activation="sigmoid"))
model.summary()
```

```python
model.compile(loss="binary_crossentropy",
              optimizer="adam", 
              metrics=["accuracy"])
```

```python
train_history = model.fit(traindf_pad, 
                  traindf["sentiment"],
                  epochs=3,
                  batch_size=200,
                  verbose=2,
                  validation_split=0.1)
```

```python
model.evaluate(testdf_pad, testdf["sentiment"])
```

```python
from keras.models import Model
partial = Model(model.input, model.layers[0].output)
partial.summary()
```

```python
result = partial.predict(testdf_pad)
result.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout, Embedding, SimpleRNN
model = Sequential()
model.add(Embedding(input_dim=2000, output_dim=32, input_length=100))
model.add(SimpleRNN(16))
model.add(Dense(units=256, activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(units=1, activation="sigmoid"))
model.summary()
```

```python
model.compile(loss="binary_crossentropy",
              optimizer="adam", 
              metrics=["accuracy"])
train_history = model.fit(traindf_pad, 
                  traindf["sentiment"],
                  epochs=3,
                  batch_size=200,
                  verbose=2,
                  validation_split=0.1)
```

```python
model.evaluate(testdf_pad, testdf["sentiment"])
```
